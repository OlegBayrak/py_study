```python
#Let's decide what is the source of the input - file or terminal?
import matplotlib.pyplot as plt

def write_header():
    print('\t' + "Name" + '\t' + "Age" + '\t' + "Score")

input_source = input("Please enter the source of the input (file or terminal): ")
continue_input = True
if input_source == "terminal" or input_source == "t" :
    write_header()
    while continue_input:
        name = input("Please enter your name: ")
        age = input("Please enter your age: ")
        score = input("Please enter your score: ")
        print('\t' + name.title() + '\t' + str(round(float(age))) + '\t' + score)
        continue_input = input("Do you want to continue? (yes/no): ")
        if continue_input.lower() == "no" or continue_input.lower() == "n":
            continue_input = False
        elif continue_input.lower() == "yes" or continue_input.lower() == "y":
            continue_input = True
        else: 
            print("Invalid input. Exiting the program.")
            break
elif input_source.lower() == "file" or input_source.lower() == "f": 
    write_header()
    # Read the data from the file
    with open('data.txt', 'r') as file:
        lines = file.readlines()

    # Step 2: Parse the data into a list of tuples
    data = [line.strip().split() for line in lines]
    print(data)
    names = [item[0] for item in data]
    print(names)
    ages = [int(item[1]) for item in data]
    print(ages)
    scores = [int(item[2]) for item in data]
    print(scores)

    # Step 3: Create a bar chart for ages
    plt.figure(figsize=(10, 5))
    plt.bar(names, ages, color='blue', alpha=0.7, label='Age')
    plt.xlabel('Names')
    plt.ylabel('Age')
    plt.title('Age of Individuals')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Step 4: Create a bar chart for scores
    plt.figure(figsize=(10, 5))
    plt.bar(names, scores, color='green', alpha=0.7, label='Score')
    plt.xlabel('Names')
    plt.ylabel('Score')
    plt.title('Scores of Individuals')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

else :
    print("Invalid input. Exiting the program.")




```
